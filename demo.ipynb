### Normalize text

```python
from whisper_normalization import EnglishTextNormalizer

# Load the English text normalizer
normalizer = EnglishTextNormalizer()

# Normalize a sample text
normalizer("Hello, world!")
```

### Normalize numbers

```python
from whisper_normalization import EnglishNumberNormalizer

normalizer = EnglishNumberNormalizer()

normalizer("I have 100 dollars")
```
